```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
label2id = {"sadness": 0, "joy": 1, "love": 2,
            "anger": 3, "fear": 4, "surprise": 5}
id2label = {
    "0": "sadness",
    "1": "joy",
    "2": "love",
    "3": "anger",
    "4": "fear",
    "5": "surprise",
}
emotion_list = list(label2id.keys())

bert = pd.read_csv(
    "./bert.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)
las_losowy_svd = pd.read_csv(
    "./las_losowy_svd.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)

las_losowy_tfidf = pd.read_csv(
    "./las_losowy_tfidf.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)

las_losowy_w2vec = pd.read_csv(
    "./las_losowy_w2v.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)

logistic_regression_svd = pd.read_csv(
    "./logistic_regression_svd.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)

logistic_regression_tfidf = pd.read_csv(
    "./logistic_regression_tfidf.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)

logistic_regression_w2vec = pd.read_csv(
    "./logistic_regression_w2v.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)

svc_svd = pd.read_csv(
    "./svc_svd.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)
svc_tfidf = pd.read_csv(
    "./svc_tfidf.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)
svc_w2vec = pd.read_csv(
    "./svc_w2v.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)
lstm_glove = pd.read_csv(
    "./lstm_glove.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)

lstm_no_glove = pd.read_csv(
    "./lstm_no_glove.csv",
    usecols=["precision", "recall", "f1-score", "name"],
    index_col="name",
)
```

```python
def get_f1_score(df: pd.DataFrame):
    return df["f1-score"][8]


def get_recall_score(df: pd.DataFrame):
    return df["recall"][8]


def get_precision_score(df: pd.DataFrame):
    return df["precision"][8]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
models = {
    "las losowy (TFIDF)": get_precision_score(las_losowy_tfidf),
    "las losowy (SVD)": get_precision_score(las_losowy_svd),
    "las losowy (word2vec)": get_precision_score(las_losowy_w2vec),
    "regresja logistyczna (TFIDF)": get_precision_score(logistic_regression_tfidf),
    "regresja logistyczna (SVD)": get_precision_score(logistic_regression_svd),
    "regresja logistyczna (word2vec)": get_precision_score(logistic_regression_w2vec),
    "SVC (TFIDF)": get_precision_score(svc_tfidf),
    "SVC (SVD)": get_precision_score(svc_svd),
    "SVC (word2vec)": get_precision_score(svc_w2vec),
    "BIdirectional LSTM": get_precision_score(lstm_no_glove),
    "Bert": get_precision_score(bert),
}

models = dict(sorted(models.items(), key=lambda item: item[1]))

# Define distinct colors for each model
colors = plt.cm.viridis(np.linspace(0, 1, len(models)))

# Create a bar plot
fig, ax = plt.subplots(figsize=(12, 6))

bars = ax.bar(range(0, len(models.keys())), models.values(), color=colors)

# Add tick labels for each model
plt.xticks(range(len(models)), range(0, len(models)), ha="right")

# Add legend for each bar
idx = 0
for bar, model in zip(bars, models):
    # Include the number in the legend
    bar.set_label(f"{idx}-{model}: {models[model]}")
    idx += 1

plt.xlabel("Modele")
plt.ylabel("Precision")
plt.title("Zestawienie metryki precision midzy modelami")
plt.ylim(0, 1)
plt.legend(loc="best", bbox_to_anchor=(1, 1))
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
